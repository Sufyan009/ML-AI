```python
import numpy as np

def gradient_descent(x, y): 
    m_cur = b_cur = 0
    iterations = 20
    n = len(x)
    learning_rate = 0.01

    for i in range(iterations):
        y_pred = m_cur * x + b_cur
        # cost (y - y_pred)
        cost = (1/n) * sum([val**2 for val in (y - y_pred)])
        md = -(2/n) * sum(x * (y - y_pred))
        bd = -(2/n) * sum(y - y_pred)        
        m_cur = m_cur - learning_rate * md
        b_cur = b_cur - learning_rate * bd  
        
        print("m {}, b {}, cost {} iteration {}".format(m_cur, b_cur, cost, i))

# Sample data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 7, 9, 11, 13, 15])

gradient_descent(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y, learning_rate=0.001, iterations=50): 
    m_cur = b_cur = 0
    n = len(x)
    cost_history = []

    plt.figure(figsize=(14, 6))

    for i in range(iterations):
        y_pred = m_cur * x + b_cur
        cost = (1/n) * sum([val**2 for val in (y - y_pred)])
        cost_history.append(cost)

        # Update parameters
        md = -(2/n) * sum(x * (y - y_pred))  # Gradient w.r.t m
        bd = -(2/n) * sum(y - y_pred)        # Gradient w.r.t b
        m_cur = m_cur - learning_rate * md
        b_cur = b_cur - learning_rate * bd

        # Plot the line of best fit after every few iterations to visualize progress
        if i % (iterations // 10) == 0 or i == iterations - 1:  # Plot 10 snapshots + final iteration
            plt.subplot(1, 2, 1)
            plt.scatter(x, y, color='blue')  # Actual points
            plt.plot(x, y_pred, color='red', label=f"Iteration {i}, Cost: {cost:.2f}")
            plt.xlabel("X")
            plt.ylabel("Y")
            plt.legend(loc="best")
            plt.title("Gradient Descent Optimization")

    # Plot cost reduction over iterations
    plt.subplot(1, 2, 2)
    plt.plot(range(iterations), cost_history, color='purple')
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    plt.title("Cost Reduction Over Iterations")

    plt.tight_layout()
    plt.show()

    print("Final values of m and b:", m_cur, b_cur)
    print("Final cost:", cost_history[-1])

# Sample data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 7, 9, 11, 13, 15])

# Run gradient descent with visualization
gradient_descent(x, y, learning_rate=0.01, iterations=50)
```
