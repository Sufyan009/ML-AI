# Bar Plot

```python
import matplotlib.pyplot as plt
x = ['Python', 'Java-Script', 'C++', 'PHP']
y = [70, 80, 90, 60]
plt.xlabel("Programming Languages", fontsize=15)
plt.ylabel("Numbers", fontsize=15)
plt.title("Bar Plot of Programming Languages", fontsize=15)
color = ['y', 'b', 'g', 'black']
plt.bar(x, y, width = 0.5, color = color, label = 'Devsinc')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv("Automobile.csv")
auto.head()
```

```python
sns.histplot(auto['engine_size'])
plt.show()
```

```python
sns.displot(auto['city_mpg'], kde=True, rug=True)
plt.show()
```

```python
sns.displot(auto['normalized_losses'], kde=True, rug=True)
plt.show()
```

```python
sns.jointplot(x='engine_size', y='horsepower', data=auto, kind='reg')
plt.show()
```

```python
sns.jointplot(x='engine_size', y='horsepower', data=auto, kind='kde')
plt.show()
```

```python
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']])
```

```python
sns.countplot(auto['body_style'])
```
