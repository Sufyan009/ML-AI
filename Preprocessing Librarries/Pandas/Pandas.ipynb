**Time Series Analysis using Apple Historical Data set - Pandas Time Series Analysis Part 1: DatetimeIndex and Resample**

```python
import pandas as pd
df = pd.read_csv("apple.csv", parse_dates=["Date"], index_col="Date")
df.head()
```

```python
# Accessing the first element using .iloc (by position)
print(type(df.Close.iloc[0]))
print(type(df.Volume.iloc[0]))
print(type(df.Open.iloc[0]))
print(type(df.High.iloc[0]))
print(type(df.Low.iloc[0]))
```

```python
print(df.index)
```

```python
print(df.columns)
```

```python
# Accessing data for January 2017
january_data = df.loc['2017-01']
january_data
```

```python
%matplotlib inline

# Use a raw string (r'...') to avoid escape sequence issues
df['Close'] = df['Close'].replace({r'\$': '', ' ': ''}, regex=True)

# Convert the column to numeric values
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')

# Resample by month-end ('ME') and calculate the mean
monthly_mean = df['Close'].resample('YE').mean().plot(kind="bar")

# Display the result
print(monthly_mean)
```

# Window Functions with Pandas

```python
import pandas as pd
import numpy as np

# Create a DataFrame with 10 rows of data
np.random.seed(0)
data = np.random.randn(10)
date = pd.date_range('20230101', periods=10)
df = pd.DataFrame(data, columns=['value'], index=date)
df.head(10)
```

### Rolling Windows
Rolling window performs calculations based on the number of specified elements in the window.


```python
df['rolling_mean'] = df['value'].rolling(window=3).mean()
print(df)
```

### Expanding Window
The expanding window performs calculations from the beginning to the current element, so its size increases as the index moves.

```python
df['value_expanding_mean_1'] = df['value'].expanding(min_periods=1).mean()
df
```

### Exponential Weighted Window
The exponential weighted window gives greater weight to newer values.

```python
df['Value_EWM_Mean_3'] = df['value'].ewm(span=3).mean()
df.head()
```
